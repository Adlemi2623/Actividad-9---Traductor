# TRADUCCIÓN DE PALABRAS CON DICC.

## Crea una lista de traducciones, busca una palabra y la muestra en pantalla.

```python
def translate(word):
    return dic.get(word, "no hallada")

word = "Casa"
print(translate(word))
```

## TRADUCCIÓN DE PALABRAS CON DICC.

## Usa un diccionario con traducciones, una función para buscar palabras y un bucle para imprimirlas todas.

```python
dic = {"hola": "hello", "Piña": "Pineapple", "Palabaras": "Words", "Sol": "Soon", "Padre": "Dad", "Mama": "Mother", "Flores": "Flowers", "Casa": "House", "Corazon": "Heart", "Cara": "Face"}

def translate(word):
    return dic.get(word, "no hallada")

for word in dic:
    print(translate(word))
```
